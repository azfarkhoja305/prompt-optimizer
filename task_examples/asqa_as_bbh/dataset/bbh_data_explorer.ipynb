```python
import pandas as pd
from task_examples.string_pager_widget import StringPager
```

```python
TRAIN_FILE = "train.csv"
VALID_FILE = "valid.csv"

train_data = pd.read_csv(TRAIN_FILE)
valid_data = pd.read_csv(VALID_FILE)

print(train_data.shape)
print(valid_data.shape)
```

```python
train_data.head()
```

```python
def data_df_to_list_str(df):
    output_list = []
    for i, (_, row) in enumerate(df.iterrows()):
        output = f"Sample: {i + 1}\n\n"
        output += f"Input:\n{row['input']}\n\n"
        output += f"Target: {row['target']}"
        output_list.append(output)
    return output_list
```

```python
train_data_str = data_df_to_list_str(train_data.head())

StringPager(train_data_str).display()
```

```python
valid_data_str = data_df_to_list_str(valid_data.head())

StringPager(valid_data_str).display()
```
